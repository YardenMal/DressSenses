```python
'''making the system interactive and allow the user uploading images. In the system, there will be 3 'upload' buttons:
for shirts image, bottoms image and shoes image'''

import ipywidgets as widgets
from IPython.display import display


main_display = widgets.Output()

def show_it(inputs):
    with main_display:
        main_display.clear_output()
        image_name = list(inputs['new'].keys())[-1]
        display(image_name)
        return image_name

upload = widgets.FileUpload(
    accept='.jpg',
    multiple=False
)
print("Please Upload an image of your items")
upload.observe(show_it, names='value')
upload
```
